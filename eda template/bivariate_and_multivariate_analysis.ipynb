## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic, x='Sex',y='Age')
```

```python
sns.boxplot(titanic, x='Sex',y='Age', hue="Survived")
```

## 4. Distplot (Numerical - Categorical)

```python
sns.kdeplot(data=titanic[titanic['Survived']==0], x="Age")
sns.kdeplot(data=titanic[titanic['Survived']==1], x="Age")
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic['Survived'] = pd.to_numeric(titanic['Survived'])
```

```python
titanic.info()
```

```python
titanic['Pclass'] = pd.to_numeric(titanic['Pclass'], errors='coerce')
titanic['Survived'] = pd.to_numeric(titanic['Survived'], errors='coerce')
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
titanic.groupby('Sex')['Survived'].mean() * 100
```

```python
titanic.groupby('Embarked')['Survived'].mean() * 100
```
